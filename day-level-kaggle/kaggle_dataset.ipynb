```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as plticker
```

```python
data = pd.read_csv("./time_series_covid_19_confirmed.csv")
```

```python
data.head
```

```python
print(data.columns)
```

```python
data.set_index("Country/Region")
```

```python
data.rename(columns = {'Country/Region' : 'Country'}, inplace = True)
```

```python
data.columns
```

```python
data = data[data.Country == "US"]
```

```python
data.head
```

```python
data.rename(columns = {'Province/State' : 'State'}, inplace = True)
```

```python
data.head
```

```python
data_md = data[data.State == "Maryland"]
```

```python
data_md.head
```

```python
data_md.columns
```

```python
data_md_dates = data_md.columns[4::]
```

```python
data_md_dates
```

```python
data_md.set_index("State")
```

```python
reports = []
for i in data_md_dates:
    reports.append(data_md.iloc[0][i])
```

```python
print(reports)
```

```python
ax = plt.gca() # get the current axes

for label in ax.xaxis.get_ticklabels():
    label.set_rotation(30)
    
loc = plticker.MultipleLocator(base=10.0) # this locator puts ticks at regular intervals
ax.xaxis.set_major_locator(loc)
plt.scatter(data_md_dates, reports)
plt.title("Reported cases in Maryland")
plt.show
```

```python
ax = plt.gca() # get the current axes

for label in ax.xaxis.get_ticklabels():
    label.set_rotation(30)
    
stroff = 40
loc = plticker.MultipleLocator(base=2.0) # this locator puts ticks at regular intervals
ax.xaxis.set_major_locator(loc)
plt.plot(data.columns[stroff::], data_md.iloc[0][stroff::], label="MD")
plt.plot(data.columns[stroff::], data.iloc[23][stroff::],color='red', label="DC" )
plt.plot(data.columns[stroff::], data.iloc[19][stroff::],color='yellow', label="VA" ) 
plt.title("Reported cases in the DMV")
plt.xlabel("Dates")
plt.ylabel("# Reported Cases")
plt.legend()
plt.show
```

```python
dc_values = pd.Series(data.iloc[23][stroff::])
dc_sum = dc_values.sum()
```

```python
dc_sum
```

```python
va_values = pd.Series(data.iloc[19][stroff::])
va_sum = va_values.sum()
```

```python
labels = 'DC', 'MD', 'VA'
vals = [dc_sum, md_sum, va_sum]
explode = (0, 0.1,0)
plt.pie(vals, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True)
# plt.axis('equal')
plt.title("Distribution of Cases in the DMV as of 3/14")
plt.show
```
