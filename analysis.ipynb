```python
import matplotlib.pyplot as plt
import matplotlib.ticker as plticker

import numpy as np
```

```python
import pandas as pd
```

```python
data = pd.read_csv("./COVID-19-geographic-disbtribution-worldwide-2020-03-14_1.csv")
```

```python
data.head()
```

```python
data.tail
```

```python
print(data.columns)
```

```python
data.set_index("CountryExp")
```

```python
data = data[data.CountryExp == "United States of America"]
```

```python
dates = data.DateRep
```

```python
print(type(dates))
```

```python
dates.iloc[::-1]
```

```python
cases = data.NewConfCases
```

```python
dates = dates.iloc[::-1]
```

```python
cases = cases.iloc[::-1]
```

```python
print(dates.size)
```

```python
ax = plt.gca() # get the current axes
plt.plot(dates,cases)
plt.title("Unique cases in the United States daily")

for label in ax.xaxis.get_ticklabels():
    label.set_rotation(30)
    
loc = plticker.MultipleLocator(base=10.0) # this locator puts ticks at regular intervals
ax.xaxis.set_major_locator(loc)
plt.show()
plt.show
```

```python
deaths = data.NewDeaths
deaths = deaths.iloc[::-1]
```

```python
ax = plt.gca()
plt.plot(dates,deaths)
plt.title("Reported deaths in the United States daily")

for label in ax.xaxis.get_ticklabels():
    label.set_rotation(30)
    
loc = plticker.MultipleLocator(base=10.0)
ax.xaxis.set_major_locator(loc)
plt.show
```
